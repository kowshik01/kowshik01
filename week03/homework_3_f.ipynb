Write a rule based predictor for a given matrix

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

dataframe = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/ds602/iris.csv')
dataframe.info()
```

```python
dataframe.species.value_counts()
```

```python
# Split the DataFrame into X (features) and y (target)
X = dataframe.iloc[:, :-1]
y = dataframe.iloc[:, -1]

# Print X (features)
print(X)

# Print y (target)
print(y)
```

```python
from sklearn.decomposition import PCA

# Instantiate PCA with 2 components
pca = PCA(n_components=2)

# Fit PCA to the data and transform X
X_pca = pca.fit_transform(X)

# Print the transformed data
print(X_pca)
```

```python
#Performing rule based prediction using PCA model
def predict_one(sepal_length, sepal_width, petal_length, petal_width):
   # Transform the input using the PCA model
    pca_input = pca.transform([[sepal_length, sepal_width, petal_length, petal_width]])
    
    # Set the decision thresholds
    threshold_1 = 1.5
    threshold_2 = 0.5
    
    # Apply rule-based prediction
    if pca_input[0, 0] <= threshold_1:
        return "Iris-setosa"
    elif pca_input[0, 0] <= threshold_2:
        return "Iris-versicolor"
    else:
        return "Iris-virginica"
```

```python
#Calculating accuracy
predictions = pd.Series([ predict_one(*line) for line in dataframe.iloc[:, :4].to_numpy()])
accuracy = sum(dataframe.species == predictions) / len(dataframe)
print(f"Accuracy: {accuracy}")
```
